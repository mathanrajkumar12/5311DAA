```
import time
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import Polynomial

# Original function given
def original_function(n):
    x = 1
    for i in range(n):
        for j in range(n):
            x = x + 1
    return x

#for various n values
ns = np.array(range(1, 1001, 50))  # Test from n = 1 to 1000
original_time = []

for n in ns:
    start = time.time()
    original_function(n)
    end = time.time()
    original_time.append(end - start)

# Fit a polynomial to the timings (quadratic curve)
coeffs = np.polyfit(ns, original_time, 2)
p = Polynomial(coeffs[::-1])  # Polynomial object

# Plot time vs n
plt.plot(ns, original_time, 'bo-', label='Original Function')
plt.plot(ns, p(ns), 'r-', label='Fitted Polynomial Curve')
plt.xlabel('n')
plt.ylabel('Time (s)')
plt.title('Time vs n for Original Function')
plt.grid(True)
plt.legend()
plt.show()
```

```
# Defining upper and lower bounds
upper_bound = lambda n: 1.2 * p(n)
lower_bound = lambda n: 0.8 * p(n)

# Plot with upper and lower bounds
plt.plot(ns, original_time, 'bo-', label='Original Function')
plt.plot(ns, p(ns), 'r-', label='Fitted Polynomial Curve')
plt.plot(ns, upper_bound(ns), 'p--', label='Upper Bound')
plt.plot(ns, lower_bound(ns), 'v--', label='Lower Bound')
plt.xlabel('n')
plt.ylabel('Time (s)')
plt.title('Time vs n with Upper and Lower Bounds')
plt.legend()
plt.grid(True)
plt.show()
```

```
# Adding a vertical line at n_0
n_0 = 200  # Eyeballed based on the plot
plt.plot(ns, original_time, 'bo-', label='Original Function')
plt.plot(ns, p(ns), 'r-', label='Fitted Polynomial Curve')
plt.axvline(x=n_0, color='orange', linestyle='--', label='n_0 = {}'.format(n_0))
plt.xlabel('n')
plt.ylabel('Time (s)')
plt.title('Time vs n with n_0 Indicated')
plt.legend()
plt.grid(True)
plt.show()
```
